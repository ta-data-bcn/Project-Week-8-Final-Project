```python
import pandas as pd
import numpy as np
import re
```

# Barcelona

```python
bcn=pd.read_csv('../data/barcelona_data.csv')
```

```python
bcn.drop(bcn[bcn['Price']=='€\xa0666666'].index, inplace=True)
bcn.drop(bcn[bcn['Pets']=='0'].index, inplace=True)
bcn_list=bcn['Pets'].tolist()
symbol = '\n'
bcn_list_clean=[item.translate(symbol).strip() for item in bcn_list]
bcn.drop(['Pets'],axis=1, inplace=True)
bcn['Pets'] = bcn_list_clean
bcn.drop(['Unnamed: 0'],axis=1, inplace=True)
bcn_prices=bcn['Price'].tolist()

clean_prices=[]
for i in bcn_prices:
    clean_prices.append(re.findall(r"\b[0-9]+\b",i)[0])
    
bcn.drop(['Price'],axis=1, inplace=True)
bcn['Price'] = clean_prices
bcn = bcn[['Name', 'Score', 'Price', 'Pets', 'Link']]
```

```python
bcn.insert(0, 'City', 'Barcelona')
bcn
```
