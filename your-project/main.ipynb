# Defining functions

### Defining a function to get only the interesting data:

As every column is on different time lengths, we need to remove all the NaN produced after the impact.

```python
def cut_data(data):
    new_data = data[~data.isna()][2:].astype('int64')
    # Taking only from row 2 because 0 and 1 are descriptive.
    return new_data

data_str = cut_data(data['B.4'])
data_str.max()
```

### Defining a function to set the mean to 0

We need to remove the offset created by the lecture.
We substract the mean from the number to have data gathered around value 0.

```python
# We could set the mean to 1026, as it seems to be the starting value for almost every lecture. 

def norm(data):
    data_mean = int(data.mean())
#     Because the data is full of integers, we don't want to convert the dataset to a float series.
    new_data = []
    for i in range(len(data)):
        new_point = data.iloc[i] - data_mean
        new_data.append(new_point)
    return new_data
    
# data_norm = norm(data_str)
# max(data_norm)
```

### Defining a function to remove small noise

There seems to be static noise in the data. We can define the upper and lower limits to remove it.

```python
def small_noise(data, lower, upper):
    new_data = [x if (x>upper or x<lower) else 0 for x in data]
    return pd.DataFrame(new_data)

# data_non_noise = small_noise(data_norm, -10, 10)
# data_non_noise.max()
```

### Defining a function to smooth the points

To smooth the points, we make a rolling mean from the data.

```python
def rolling(data, smoother):
    new_data = data.rolling(window = smoother).sum()
    return pd.DataFrame(new_data)

# smooth = 3
# data_rolled = rolling(data_non_noise, smooth)
# data_rolled
```

# Creating a new DataFrame with the hitting point of every strike

#### We need to drop the first column, as it only contains headers

```python
data.drop(columns = 'Technique', inplace = True)
```

```python
lower, upper, smooth, min_time, max_time = -8, 5, 3, 30, 50

clean_data = pd.DataFrame()
j = 0

for column in data.columns:
    name = data[column][0] + '-' + column[0] + '-' + data[column][1]
#     The name will take the person's ID, the class of the strike and the number of trial

    clean_ = rolling(small_noise(norm(cut_data(data.iloc[:,j])), lower, upper), smooth)
#     We just run all the little functions to clean the data
    
    pos_max = clean_[clean_[0] == clean_[0].max()].index[0]
#     We find which is the maximum of the dataset

    temp_list = clean_.iloc[pos_max - min_time: pos_max + max_time]
#     We determine the range of the data we want
    
    clean_data[name] = temp_list[0].reset_index(drop=True)
#     We add it to a new dataframe

    j += 1
# clean_data = clean_data.transpose()
# # We transpose the data in order to access them directly.

clean_data.to_csv('data/clean_data.csv', sep = ',')
# Save the new DataFrame into a CSV file for further manipulations.
```

We can make means of every kick, by participant.

```python
def mean_of_kick(data, value):
    test_df = pd.DataFrame()

    for i in data.columns:
        if i[:4] == value:
            test_df[i] = data.loc[:, i].reset_index(drop=True)

    new_mean = test_df.mean(axis=1)
    
    return new_mean


data_means = pd.DataFrame()
for i in new_list_of_person_per_kick:
    data_means[i] = mean_of_kick(added_data, i)
    
# data_means
```

We can also make maxs of every kick, by participant.

```python
def max_of_kick(data, value):
    test_df = pd.DataFrame()

    for i in data.columns:
        if i[:4] == value:
            test_df[i] = data.loc[:, i].reset_index(drop=True)

    new_max = test_df.max(axis=1)
    
    return new_max


data_maxs = pd.DataFrame()
for i in new_list_of_person_per_kick:
    data_maxs[i] = max_of_kick(added_data, i)
    
dict_max = dict(data_maxs.iloc[30])
# dict_max
```
