```python
#for x in range(len(pd)):
def get_track_features(df):
    track_features = pd.DataFrame()
    for x in range(len(df)):
        track_id = df['track.id'][x]
        if x%10 == 0:
            time.sleep(np.random.uniform(sleep_min, sleep_max))
        try:
            track_data = sp.audio_features(track_id)
        except:
            oauth2 = util.oauth2.SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
            token = util.prompt_for_user_token(username, scope=scope, client_id='22e782a724264812b124038bbf783396', client_secret='6219554f45764f2cbfe13306ec476f10', redirect_uri='https://localhost:8080')
            sp = spotipy.Spotify(auth=token)
            track_data = sp.audio_features(track_id)
        data = (json_normalize(track_data))
        track_features = track_features.append(data,ignore_index=True)
        save_tracks = track_features.to_csv(username+'_track_features.csv')
        print(f'{track_id}')
    return track_features
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import seaborn as sns
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial import distance

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances_argmin_min
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(components[:, 0], components[:, 1], components[:, 2],s=60)
Axes3D.mouse_init(ax)
```

```python
kmeans = KMeans(n_clusters=3).fit(components)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(components)
print(labels)
# Getting the cluster centers
C = kmeans.cluster_centers_
print(C)
colors=['tomato','lightskyblue','mediumaquamarine']
to_color=[]
for row in labels:
    to_color.append(colors[row])

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(components[:, 0], components[:, 1], components[:, 2], c=to_color,s=80)
ax.scatter(C[:, 0], C[:, 1],C[:, 2], marker='H', c='black', s=500)
ax.scatter(C[:, 0], C[:, 1],C[:, 2], marker='H', c=colors, s=400)
```

```python

closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, components)
closest
```

```python
songs_names=songs['track.album.name'].values
for row in closest:
    print(songs_names[row])
```

```python
import math
def distance_cluster(row):
    cluster_c1 = C[songs_cluster.loc[row]['cluster']][0]
    cluster_c2 = C[songs_cluster.loc[row]['cluster']][1]
    cluster_c3 = C[songs_cluster.loc[row]['cluster']][2]
    song_c1 = songs_cluster.loc[row]['C1']
    song_c2 = songs_cluster.loc[row]['C2']
    song_c3 = songs_cluster.loc[row]['C3']
    cluster = (cluster_c1,cluster_c2,cluster_c3)
    song = (song_c1,song_c2,song_c3)

    dst_cluster = distance.euclidean(cluster, song)

    return(dst_cluster)
```

```python
songs_cluster['distance_cluster'] = songs_cluster.index.to_series().apply(distance_cluster)
```

```python
cluster_relevance = songs_cluster['cluster'].value_counts()
songs_cluster['cluster'].value_counts()
```

```python
num_of_clusters = songs_cluster['cluster'].value_counts()
num_of_clusters.count()
```

```python
list_val= []
for x in range(num_of_clusters.count()):
    list_val.append('df_'+str(x))
    songs_cluster['cluster'].count()
    mask = songs_cluster['cluster'] == x
    list_val[x] = songs_cluster[mask]
    temp = pd.DataFrame(list_val[x])
    temp.to_csv(username + '_cluster_gigante_' + str(x) + '.csv')
```

```python
cl0_gigante = pd.read_csv(username + '_cluster_gigante_0.csv').sort_values(['distance_cluster'], ascending=True)
cl0_gigante.head(10)
```

```python
cl0 = pd.read_csv(username + '_cluster_0.csv').sort_values(['distance_cluster'], ascending=True)
cl0_describe = cl0.describe()
```

```python
cl0_describe
```

```python
cl0_dance_q1 = cl0_describe['Danceability']['min']
cl0_dance_q3 = cl0_describe['Danceability']['max']
cl0_ene_q1 = cl0_describe['Energy']['min']
cl0_ene_q3 = cl0_describe['Energy']['max']
cl0_lou_q1 = cl0_describe['Loudness']['min']
cl0_lou_q3 = cl0_describe['Loudness']['max']
cl0_spe_q1 = cl0_describe['Speechiness']['min']
cl0_spe_q3 = cl0_describe['Speechiness']['max']
cl0_aco_q1 = cl0_describe['Acousticness']['min']
cl0_aco_q3 = cl0_describe['Acousticness']['max']
cl0_ins_q1 = cl0_describe['Instrumentalness']['min']
cl0_ins_q3 = cl0_describe['Instrumentalness']['max']
cl0_liv_q1 = cl0_describe['Liveness']['min']
cl0_liv_q3 = cl0_describe['Liveness']['max']
cl0_val_q1 = cl0_describe['Valence']['min']
cl0_val_q3 = cl0_describe['Valence']['max']
cl0_tem_q1 = cl0_describe['Tempo']['min']
cl0_tem_q3 = cl0_describe['Tempo']['max']
```

```python
def is_inside_higher_cluster(row):
    if cl0_dance_q1 <= cl0_gigante['tracks_features.danceability'][row] <= cl0_dance_q3:
        if cl0_ene_q1 <= cl0_gigante['tracks_features.energy'][row] <= cl0_ene_q3:
            if cl0_lou_q1 <= cl0_gigante['tracks_features.loudness'][row] <= cl0_lou_q3:
                if cl0_spe_q1 <= cl0_gigante['tracks_features.speechiness'][row] <= cl0_spe_q3:
                    if cl0_aco_q1 <= cl0_gigante['tracks_features.acousticness'][row] <= cl0_aco_q3:
                        if cl0_ins_q1 <= cl0_gigante['tracks_features.instrumentalness'][row] <= cl0_ins_q3:
                            if cl0_liv_q1 <= cl0_gigante['tracks_features.liveness'][row] <= cl0_liv_q3:
                                if cl0_val_q1 <= cl0_gigante['tracks_features.valence'][row] <= cl0_val_q3:
                                    if cl0_tem_q1 <= cl0_gigante['tracks_features.tempo'][row] <= cl0_tem_q3:
                                        return 1
    else:
        return 0
```

```python
cl0_gigante['inside'] = 0
cl0_gigante['inside'] = cl0_gigante.index.to_series().apply(is_inside_higher_cluster)
```
